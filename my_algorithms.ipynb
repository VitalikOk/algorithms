```python
#Задача на программирование: покрыть отрезки точками

# По данным nn отрезкам необходимо найти множество точек минимального размера, для которого каждый из отрезков содержит хотя 
# бы одну из точек.

# В первой строке дано число 1 <= n <= 100 отрезков. Каждая из последующих nn строк содержит
# по два числа 0 <= l <= r <= 10^9 , задающих начало и конец отрезка. 
# Выведите оптимальное число mm точек и сами mm точек. Если таких множеств точек несколько, выведите любое из них.


import numpy as np


def get_dots(cut_count, dots):
    """
    Функция подготавливант данные и сортирует отрезки по возрастанию
    конечной координаты
    затем повторяет оптимальный шаг до окончания всех отрезков
    """
    data_type = [('begin', int), ('end', int)]
    cuts = np.array(dots, dtype = data_type)
    cuts.sort(order = 'end')
    dots = []
    
    
    def get_dot_del_cuts(cuts):
        """
        Функция оптимального шага
        возыращает конечную координату первого отрезка
        И отрезки на которын эта координата не попадает
        """
        first_dot = cuts[0]['end']
        cut_count = len(cuts)           
        for i in range(1, cut_count):          
            if cuts[i]['begin'] > first_dot:
                i-=1
                break           
                
        cuts = cuts[i+1:cut_count]               
        return first_dot, cuts
    
    
    while len(cuts) > 0:
        cur_dot, cuts = get_dot_del_cuts(cuts)
        dots.append(cur_dot)
    
    return dots


def print_result(cut_count, data):
    dots = get_dots(cut_count, data)
    print(len(dots))
    str_dots = '' 
    for dot in dots:
        str_dots += str(dot) + ' '        
    print(str_dots[:-1])


cut_count = 4
data = [(4,7),
        (1,3),
        (2,5),
        (5,6),
       ]

print_result(cut_count, data)


print('====')
cut_count = 3
data = [
        (1,3),
        (2,5),
        (3,6),
       ]
print_result(cut_count, data)

print('====')
cut_count = 5
data = [
       (5, 6),
       (4, 7),
       (3, 8),
       (2, 9),
       (1, 10),
       ]

print_result(cut_count, data)
```

```python
#Задача на программирование: непрерывный рюкзак

#Первая строка содержит количество предметов 1 <= n <= 10^3
#и вместимость рюкзака 0 <= W <= 2⋅10^6
#Каждая из следующих nn строк задаёт стоимость 0 <= ci <= 2⋅10^6
#и объём 0 < wi <= 2*10^60 предмета (n, W, ci, wi  — целые числа). 
# Выведите максимальную стоимость частей предметов (от каждого предмета можно отделить любую часть, 
# стоимость и объём при этом пропорционально уменьшатся), помещающихся в данный рюкзак, 
# с точностью не менее трёх знаков после запятой.

import numpy as np


def get_backpack_cost(data):
    """
    Функция подготавливант данные и сортирует товары по убыванию цены за еденицу объёма
    затем повторяет оптимальный шаг до окончания всех товаров
    """
    data_type = [('cost', int), ('volume', int), ('price', float)]
    count = data[0][0]
    backpack_vol = data[0][1]
    backpack_cost = 0
    items = np.zeros([count], dtype = data_type)
    for i in range(0,count):
        items[i]['cost'] = data[i+1][0]
        items[i]['volume'] = data[i+1][1]
        items[i]['price'] = items[i]['cost'] / items[i]['volume']
    items[::-1].sort(order = 'price')
    
    
    def get_next_item(items, backpack_vol, backpack_cost):
        """
        Функция оптимального шага
        Помещает в рюкзак самый стоящй товар
        возвращает общюю стоимость рюкзака, оствышийся объём
        И оставшиеся товары

        """
        if backpack_vol >= items[0]['volume']:
            backpack_cost += items[0]['cost'] 
            backpack_vol -= items[0]['volume']
            items = items[1:]
            return items, backpack_vol, backpack_cost
        else:
            backpack_cost += items[0]['price'] * backpack_vol
            backpack_vol = 0
            return items, backpack_vol, backpack_cost

    
    while backpack_vol > 0 and len(items) > 0:
        print(backpack_vol)
        print(items)
        items, backpack_vol, backpack_cost = get_next_item(items, backpack_vol, backpack_cost)
    
    return backpack_cost


def print_result(data):    
    print('{:.3f}'.format(get_backpack_cost(data)))


data = [(2,50),
        (60,20),
        (100,50),
        (120,30),
       ]

print_result(data)
```

```python
#Задача на программирование: различные слагаемые
#По данному числу 1 <= n <= 10^9
#найдите максимальное число k, для которого n можно представить как сумму k различных натуральных слагаемых. 
#Выведите в первой строке число k, во второй — k слагаемых.


def get_nums(n):
    nums = []     
    
    def get_next_num(nums, n): 
        if len(nums) > 0:
            next_num = nums[-1] + 1
        else:
            next_num = 1
        if (n - next_num) <= next_num:
            nums.append(n)
            n -= n
        else:
            nums.append(next_num)
            n -= next_num
        return nums, n
        
    while n > 0:
        nums, n = get_next_num(nums, n)
    
    return nums


def print_result(data):    
    nums = get_nums(data)
    str_nums = ''
    print(len(nums))
    for num in nums:
        str_nums += str(num) + ' '        
    print(str_nums[:-1])
    

for i in range(1,5): 
    print(i)
    print_result(i)
    print('=====')
```

## Задача на программирование: кодирование Хаффмана

```python
#По данной непустой строке s длины не более 10^4,
#состоящей из строчных букв латинского алфавита, постройте оптимальный беспрефиксный код. 
#В первой строке выведите количество различных букв k, встречающихся в строке,
#и размер получившейся закодированной строки. 
#В следующих k строках запишите коды букв в формате "letter: code". 
#В последней строке выведите закодированную строку.



def count_ins(s):
    """
    Функция возвоащает словарь со всеми встречающимися символами в качестве ключей
    И частоту их появления в качестве значений
    """
    ar_str = {}
    for i in range(len(s)):
        if s[i] in ar_str:
            ar_str[s[i]] += 1
        else:
            ar_str[s[i]] = 1        
#     print(ar_str)        
    k = list(ar_str.keys())
    
    H = []
    for i in range(len(k)):
        H.append([i,ar_str[k[i]]])


    return H, k 
            
def extract_min(H):
    """
    Функция находит минимум в массиве H[i][1]
    И возвращает этот минимм и массив без него
    """
    if len(H) > 0:
        min_i = 0
        min_f = H[min_i][1]
        for i in range(len(H)):
            if H[i][1] < min_f:
                min_i = i
                min_f = H[i][1]
        min_n = H[min_i]
        del H[min_i]
    else: 
        min_n=[0,0]
    return H, min_n

def build_tree(H, k):
    """
    Функция строит дерево по алгоритму кодирования Хаффмана
    """
    len_coded_s = 0
    n = len(k)
    if  n == 1:
        n = 2
    for n in range(1, n):
        H, min_i = extract_min(H)
        H, min_j = extract_min(H)    
        H.append([len(k),  min_i[1] + min_j[1]])
        len_coded_s += H[-1][1]
        k.append([min_i[0], min_j[0]])
#         print(H, k)
    return len_coded_s, k
    
def get_simb_code(simb, k, n_simb):
    """
    функция возвращает код переданного символа из дерева k
    """
    
    def get_simb_id(simb):
        """
        функция возвращает id (позицию) переданного симаолв
        """
        for i in range(0,n_simb):
            if k[i] == simb:
                return i
    
    def get_next_code(ind, code=''):
        """
        функция возвращает код Хаффмана для id переданного символа
        """
        for i in range(n_simb, len(k)):
            if k[i][0] == ind:
                code = get_next_code(i, '0'+code)
                break
            elif k[i][1] == ind:
                code = get_next_code(i, '1'+code)
                break
        return code
    
    simb_id = get_simb_id(simb)
    
    return get_next_code(simb_id)        

def get_str_code(s, k, n_simb):
    """
    функция возвращает код переданной строки из дерева k
    """
    result_str = ''
    for i in range(len(s)):
        result_str += get_simb_code(s[i], k, n_simb)
    return result_str



def print_result(s):
    """
    Функция вывода результата
    """
    H, k = count_ins(s)
    print(H, k)
    n_simb = len(k)
    len_cs, k = build_tree(H, k)

    print(f'{n_simb} {len_cs}') 

    for i in range(n_simb):
        print(f'{k[i]}: {get_simb_code(k[i], k, n_simb)}')
    
    print(get_str_code(s, k, n_simb))
    

s = 'abacabad'

print_result(s)
```

```python
# Задача на программирование: очередь с приоритетами
# Первая строка входа содержит число операций 1 ≤ n ≤ 10^5
# Каждая из последующих n строк задают операцию одного из следующих двух типов:
# Insert x, где 0 ≤ x ≤ 10 ^ 9 — целое число;
# ExtractMax.
# Первая операция добавляет число xx в очередь с приоритетами, вторая — извлекает максимальное число и выводит его.

import math

N = [255, 168, 246]

def insert(d):    
    child = len(N)
    parent = 0
    N.append(d)
    while child > 0:        
        parent = int(math.ceil(child / 2) -1)
        if parent < 0:
            parent = 0            
        if N[parent] < N[child]:
            N[parent], N[child] = N[child], N[parent]
            child = parent
        else:
            break

def move_down():
    parent, child, child_1, child_2 = 0,0,0,0
    len_n = len(N)
    while child_1 < len_n and child_2 < len_n:
        child_1 = parent*2 + 1
        child_2 = parent*2 + 2
        if child_1 < len_n:
            child = child_1
            if child_2 < len_n:
                if N[child_2] > N[child_1]:
                    child = child_2
                        
        if parent != child and N[parent] < N[child]:
            N[parent], N[child] = N[child], N[parent]
            parent = child
    
    
def extract_max():
    if len(N) > 0:
        result = N[0]
        N[0] = N[len(N)-1]    
        del N[len(N)-1]        
        move_down()
        print(result)        
                    
def do_command(command, n=None):
    if command == 'Insert':
        insert(n)
    elif command == 'ExtractMax':
        extract_max()
          
    
commands = [    
            ['Insert', 200],
            ['Insert', 10],
            ['ExtractMax', ''],
            ['Insert', 5],
            ['Insert', 500],            
            ['ExtractMax', ''],         
            ['ExtractMax',''],
            ['ExtractMax',''],
            ]


for k  in commands:
    do_command(k[0],k[1])     
```
