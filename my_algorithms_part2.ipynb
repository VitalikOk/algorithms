## Двоичный поиск

```python
"""
В первой строке даны целое число 1 ≤ n ≤ 10^5
и массив A[1...n] из n различных натуральных чисел, не превышающих 10^9, в порядке возрастания, 
во второй — целое число 1 ≤ k ≤ 10^5 и k натуральных чисел b_1.....b_k, не превышающих 10^9
Для каждого i от 1 до k необходимо вывести индекс 1 ≤ j ≤ n, для которого A[j] = b_i , или −1, если такого j нет.
"""


def get_array(str):
    arr = list(map(int, str.split(' ')))
    return arr

def find_dig(d, arr):
    l,r = 1, arr[0]
    mid = l + ((r - l) // 2)
    while (l <= mid <= r):           
        if arr[mid] > d:
            r = mid - 1
        elif arr[mid] < d:
            l = mid + 1
        else:
            return mid
        mid = l + ((r - l) // 2)
    return -1
        

def get_ind_str(A, B):
    A = get_array(A)
    B = get_array(B)
    result = ''
    for b in B[1:]:
        result += str(find_dig(b, A))+' '
    return result
    
inpt_str_1 = '6 1 5 8 12 13 23'
inpt_str_2 = '5 8 1 23 1 11'

print(get_ind_str(inpt_str_1, inpt_str_2))
```

## Сортировка слиянием

```python
"""
#Сортировка слиянием
"""

def mrg(a, b):
    c = []
    while len(a) > 0 and len(b) > 0:        
        if a[0] > b[0]:
            c.append(b[0])
            del b[0]
        else:
            c.append(a[0])
            del a[0]
    c.extend(a)
    c.extend(b)
    return c


def sort_merge(arr):    
    for i in range(len(arr)): 
        arr[i] = [arr[i]]
    cnt = 0
    while len(arr) > 1:
        if cnt+1 >= len(arr):
            cnt = 0
        arr[cnt] = mrg(arr[cnt],arr[cnt+1])
        del arr[cnt+1]
        cnt += 1
    return arr[0]
        
        
arr = [5,8,4,0,64,568,9,4,0,0,5,69,37]
print(sort_merge(arr.copy()))
arr.sort()
print(arr)
```
